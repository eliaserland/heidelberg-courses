```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Import density and potential fields.
txtname_dens = "output_density.txt"
data_density = np.loadtxt(txtname_dens)

txtname_pot = "output_potential.txt"
data_pot = np.loadtxt(txtname_pot)

#print(data_density.sum())
#print(data_pot.sum())

fsize = (15,15)

fig, axes = plt.subplots(1, 2, figsize=fsize)
axes[0].imshow(data_density)
axes[1].imshow(data_pot)
plt.show()
```

```python
# Import acceleration fields.
txtname_acc_x = "output_acc_x.txt"
data_acc_x = np.loadtxt(txtname_acc_x)

txtname_acc_y = "output_acc_y.txt"
data_acc_y = np.loadtxt(txtname_acc_y)

fsize = (15,15)

fig, axes = plt.subplots(1, 2, figsize=fsize)
axes[0].imshow(data_acc_x)
axes[1].imshow(data_acc_y)
plt.show()
```
